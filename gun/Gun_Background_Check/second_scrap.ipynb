```python
import pandas as pd
import requests as rqst
from bs4 import BeautifulSoup as bs
import numpy as np
```

```python
ULR_2019_2021_A_Income = rqst.get("https://fred.stlouisfed.org/release/tables?rid=249&eid=259515&od=2020-01-01#")

soup_19_21_AI = bs(ULR_2019_2021_A_Income.content, "html.parser")
soup_19_21_AI
```

```python
tablehead = soup_19_21_AI.thead
tablehead
```

```python
AI_Headers_19_21 = []
for x in tablehead.find_all('tr'):
    for y in x.find_all('th'):
        AI_Headers_19_21.append(y.text)
AI_Headers_19_21
```

```python
# list for thead is messed up slightly but can tell years 2020, 2021, 2019
# 2020, preceding period is 2021, year ago is 2019
```

```python
table_body = soup_19_21_AI.tbody
table_body
```

```python
TableValues_19_21 = []
for x in table_body.find_all('tr'):
    td_tags = x.find_all("td", {"class})
    #td_values = [y in td_tags]
    TableValues_19_21.append(td_tags)
TableValues_19_21    
  
```

```python
TableValues_19_21 = []
for x in table_body.find_all('tr'):
    td_tags = x.find_all("td", {"class": "fred-rls-elm-vl-td"})
    #td_values = [y.text in y in td_tags]
    TableValues_19_21.append(td_tags)
TableValues_19_21   
```

```python
#for loop with x
numbers = table_body.find_all("td", {"class": "fred-rls-elm-vl-td"})
numbers
```

```python

for x in numbers:
    numbers2 = x.text.strip()
    print(numbers2)
```

```python
listing = []
for x in numbers:
    listed = x.text.strip()
    listing.append(listed)
listing
```

```python
year2020 = []
year2019 = []
year2021 = []
select = 1
for x in numbers:
    if select==1:
        pay = x.text.strip()
        year2020.append(pay)
        select = select + 1
    if select==2:
        pay = x.text.strip()
        year2019.append(pay)
        select = select + 1
    if select==3:
        pay = x.text.strip()
        year2021.append(pay)
        select = select - 2
year2021
```

```python
splits = np.array_split(listing, 52)
splits
```

```python
#find span
A_tags = soup_19_21_AI.find_all("a")
A_tags
```

```python
#loop and strip
span_tags = soup_19_21_AI.find_all("span", {"class": "fred-rls-elm-nm"})
span_tags
```

```python
listing2 = []
for x in span_tags:
    listed2 = x.text.strip()
    listing2.append(listed2)
listing2
```

```python
df = pd.DataFrame(splits, columns = ['avg pay 2020', 'avg pay 2021', 'avg pay 2019'])
df
```

```python
df['location'] = listing2
df
```

```python
df.to_csv('avg_pay_2019_2021.csv', index=False)
```
