```python
import pandas as pd
```

```python
background_df = pd.read_csv("./Resources/permits2021-2019.csv")
background_df
```

```python
pay_df = pd.read_csv("./Resources/avg_pay_2019_2021.csv")
pay_df
```

```python
pay_df = pay_df.rename(columns={'location':'state'})
```

```python
df_pay_2019 = pay_df.drop(['avg pay 2021', 'avg pay 2020'], axis=1)
df_pay_2019 = df_pay_2019.rename(columns={'avg pay 2019':'avg pay'})
df_pay_2019
```

```python
df_pay_2019['Year'] = 2019
df_pay_2019
```

```python
df_pay_2020 = pay_df.drop(['avg pay 2021', 'avg pay 2019'], axis=1)
df_pay_2020 = df_pay_2020.rename(columns={'avg pay 2020':'avg pay'})
df_pay_2020.head()
```

```python
df_pay_2020['Year'] = 2020
df_pay_2020.head()
```

```python
df_pay_2021 = pay_df.drop(['avg pay 2020', 'avg pay 2019'], axis=1)
df_pay_2021 = df_pay_2021.rename(columns={'avg pay 2021':'avg pay'})
df_pay_2021
```

```python
df_pay_2021['Year'] = 2021
df_pay_2021
```

```python
df_pay_con = pd.concat([df_pay_2021, df_pay_2020, df_pay_2019], ignore_index=True, axis=0)
df_pay_con
```

```python
merged_df = background_df.merge(df_pay_con, on=["state", "Year"])
merged_df
```

```python
merged_df.to_csv('Resources/permits_pay_2021-2019.csv', index=False)
```
